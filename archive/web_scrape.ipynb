```python
from bs4 import BeautifulSoup
import requests
import html5lib
```

```python
url = "https://shop.tiktok.com/view/product/1731678188215699071"
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())
```

```python
table = soup.find_all('img', attrs={"class": "lazy-img"})
print(table[0]['data-src'])
```
