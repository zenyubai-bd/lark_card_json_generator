```python
import pandas as pd
```

```python
video_table = pd.read_csv("video_table.csv")
video_table = video_table[video_table["In Product Table"]==1]
```

```python
def clean_data(df):
    """
    This function aims to clean the column names to prepare the transformation to json
    """
    df = df[["handle", "Product TTS Link", "Video Link", "Product Name", "Product Category"]]
    df = df.rename(columns={
        "Product Name":     "product_name",
        "Video Link":       "video_link",
        "handle":           "creator_handle",
        "Product Category": "product_category",
        "Product TTS Link": "product_link"
    })
    return df
```

```python
video_table_selected = video_table[:5]
video_table_selected = clean_data(video_table_selected)
template_variables = video_table_selected.to_json(orient="records", force_ascii=False, indent=4)
```

```python
def reselect_videos(df, pop_row):
    """
    This function is used to delete the selected rows and regenerate json template for card generation
    Args:
        df: original dataframe
        pop_row: the row that want to be deleted
    """
    pop_row = [x for x in pop_row if x<5]
    df = df.drop(pop_row)
    df_selected = df[:5]
    df_selected = clean_data(df)
    template_variables = df_selected.to_json(orient="records", force_ascii=False, indent=4)
    return template_variables
```

```python
template_variables = reselect_videos(video_table, [2,3])
print(template_variables)
```
